```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var1 = []
random.seed(20) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
levels = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]
for _ in range(100): # why is a loop needed?
    var1 += random.sample(levels, 1) # var1.append(random.sample(levels, 1)) would also work

# Simulate var2
np.random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
var2 = np.random.uniform(0, 100, 100)
```

# Table1

```python
# Simulate table1
table1 = pd.DataFrame(var1).value_counts(sort=True)
table1 = table1.rename_axis("landuse")
table1 = table1.reset_index(name="Frequency")

print (table1)
```

Table1 is composed by two types of variables, one quanlitative nominal the "landuse" and the  other is it's frequencies, a quantitative discrete variable. So a good option to show it's distribuition can be a columns histogram.

```python


# Data from the frequency table
land_use = ['Irrigated crops', 'Managed Forest', 'Agro-Forestry system', 'Natural Forest', 'Permanent crops', 'Pasture', 'Urban', 'Shrubland']
frequency = [18, 15, 14, 13, 13, 11, 10, 6]

# Create histogram
plt.figure(figsize=(8, 6))
plt.bar(land_use, frequency, color='darkgreen')

# Adding labels and title
plt.xlabel('Land Use')
plt.ylabel('Frequency')
plt.title('Frequency of Land Use Categories')
plt.xticks(rotation=45, ha='right')  # Rotate x-labels for better readability
plt.tight_layout()

# Display the plot
plt.show()
```

# Table2

```python
# Simulate table2
table2 = pd.DataFrame(list(zip(var1, var2)), columns = ["landuse", "cover"])

print (table2)
```

Table2 is composed by two types of vaiables the first landuse is a qualitative nominal, and the second covers, is a quantitative continuos. A good representation of this variables is achieved by a box plot chart because not only show us the distribution of the covers by landuse but also allows to compare the diferences of the values of covers between landuses.

```python

# Set the style for seaborn
sns.set(style="whitegrid")

# Create a box plot
plt.figure(figsize=(5, 6))
sns.boxplot(x='landuse', y='cover', data=table2)
plt.title('Box Plot of Cover Values by Land Use')
plt.xlabel('Land Use')
plt.ylabel('Cover Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

# Table3

```python
# Simulate table3
np.random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
year = list(range(1970,2021))
temp = np.random.normal(17,2,51)
table3 = pd.DataFrame(list(zip(year, temp)), columns = ["Year", "Temperature"])

print (table3)
```

Table3 is composed by to variables, Years a quanlitative ordinal, and Temperatue a quantitative continuous. An option to discribe the changes in temperatures over the years it's a line chart.

```python


# Plotting the line chart
plt.figure(figsize=(10, 4))
plt.plot(table3['Year'], table3['Temperature'], marker='x', color='r', linestyle='-')

# Adding title and labels
plt.title('Temperature Variation in 50 Years')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.grid(True)  # Add grid for better readability

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

# Table4

```python
# Simulate table4
xx = np.array([16,21])
yy = np.array([300, 1200])
means = [xx.mean(), yy.mean()]
stds = [xx.std() / 3, yy.std() / 3]
corr = -0.7 # correlation
covs = [[stds[0]**2          , stds[0]*stds[1]*corr],
        [stds[0]*stds[1]*corr,           stds[1]**2]] # covariance matrix
table4 = pd.DataFrame(np.random.multivariate_normal(means, covs, 100), columns = ["Mean Anual Temperature", "Total Precipitation"])

print (table4)
```

Table4 is composed by two quantitative continuos values, Mean Anual Temperature ande Total Precipitation. The relation between the two can be observed in a scatter chart.

```python




# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(table4["Mean Annual Temperature"], table4["Total Precipitation"], color='green', alpha=0.5)

# Adding title and labels
plt.title("Scatter Plot of Mean Annual Temperature vs. Total Precipitation")
plt.xlabel("Mean Annual Temperature")
plt.ylabel("Total Precipitation")

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Table5

```python
# Simulate table5
col1 = pd.Series(list(range(1900,2010,10))).repeat(8)
col2 = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]*11
col3 = np.random.uniform(0, 100, 90)
table5 = pd.DataFrame(list(zip(col1, col2, col3)), columns = ["Year", "Landuse", "Cover"])

print (table5)
```

Table5 is composed by three variables, where Year and Landuse are qualitative variables ordinal and nominal respectively, and Cover is a quantitaive  continuos variable. A multi grouped bar chart can be use to do the comparison not only between years buta also between thhe covers.

```python



# Group data by year and land use category
grouped_data = table5.groupby(['Year', 'Landuse'])['Cover'].mean().unstack()

# Create the grouped bar chart
plt.figure(figsize=(14, 10))
grouped_data.plot(kind='bar', stacked=False, width=0.75)

# Adding title and labels
plt.title('Grouped Bar Chart of Cover Values Over Time by Land Use Category')
plt.xlabel('Year')
plt.ylabel('Cover Value')
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.legend(title='Land Use', bbox_to_anchor=(1.02, 1), loc='upper left', fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import random
```

```python
# Simulate var1
var1 = []
random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
levels = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]
for _ in range(100): # why is a loop needed?
    var1 += random.sample(levels, 1) # var1.append(random.sample(levels, 1)) would also work
```

var1 represents land use categories. It is generated by randomly sampling from a list of land use levels, including “Permanent crops,” “Irrigated crops,” “Managed Forest,” “Natural Forest,” “Agro-Forestry system,” “Urban,” “Pasture,” and “Shrubland.”


The loop runs 100 times, and in each iteration, a random land use level is added to the var1 list.

var1:
This variable simulates land use categories. It contains 100 randomly sampled land use types from the following list:
“Permanent crops”
“Irrigated crops”
“Managed Forest”
“Natural Forest”
“Agro-Forestry system”
“Urban”
“Pasture”
“Shrubland”
The random seed used for generating this data is 24.

```python

# Simulate var2
np.random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
var2 = np.random.uniform(0, 100, 100)
```

var2 represents a continuous variable. It is generated using NumPy’s np.random.uniform() function, which creates an array of 100 random numbers uniformly distributed between 0 and 100.

var2:
This variable simulates cover values. It contains 100 random values uniformly distributed between 0 and 100.
The random seed used for generating this data is also 24.

```python

# Simulate table1
table1 = pd.DataFrame(var1).value_counts(sort=True)
table1 = table1.rename_axis("landuse")
table1 = table1.reset_index(name="Frequency")
display(table1)

```

table1 summarizes the frequency of each land use category in var1.
It uses Pandas to create a DataFrame with columns “landuse” and “Frequency.”
The .value_counts() method counts the occurrences of each unique land use level.
The result is sorted by frequency.

table1:
This table summarizes the frequency of each land use category in var1.
It has two columns:
“landuse”: The land use category.
“Frequency”: The count of occurrences for each land use category.

```python
import matplotlib.pyplot as plt

# Line plot for table1 (Frequency of land use categories)
table1.plot(x="landuse", y="Frequency", kind="line")
plt.xlabel("Land Use Category")
plt.ylabel("Frequency")
plt.title("Frequency of Land Use Categories")
plt.show()
```

```python

# Simulate table2
table2 = pd.DataFrame(list(zip(var1, var2)), columns = ["landuse", "cover"])
display(table2)

# Note: The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together, etc. The tuple() function displays a readable version of the result - try running: print(tuple(zip(var1,var2)))

```

table2 combines var1 and var2 into a DataFrame.
It has columns “landuse” and “cover,” where “cover” corresponds to the continuous variable.

This table combines var1 (land use categories) and var2 (cover values).
It has two columns:
“landuse”: The land use category.
“cover”: The corresponding cover value.

```python
table2.plot(x="landuse", y="cover", kind="line")
plt.xlabel("Land Use Category")
plt.ylabel("Cover Value")
plt.title("Land Use Categories vs. Cover Values")
plt.show()
```

```python


# Simulate table3
np.random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
year = list(range(1970,2021))
temp = np.random.normal(17,2,51)
table3 = pd.DataFrame(list(zip(year, temp)), columns = ["Year", "Temperature"])
display(table3)
```

table3 represents temperature data over the years.

It contains columns “Year” and “Temperature.”

The temperature values are generated using a normal distribution with a mean of 17 and a standard deviation of 2.

table3:
This table simulates temperature data.
It contains two columns:
“Year”: Ranging from 1970 to 2020.
“Temperature”: Simulated temperature values drawn from a normal distribution with a mean of 17°C and a standard deviation of 2°C.

```python
table3["Temperature"].plot(kind="hist", bins=20)
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.title("Temperature Distribution")
plt.show()
```

```python

# Simulate table4
xx = np.array([16,21])
yy = np.array([300, 1200])
means = [xx.mean(), yy.mean()]
stds = [xx.std() / 3, yy.std() / 3]
corr = -0.7 # correlation
covs = [[stds[0]**2          , stds[0]*stds[1]*corr],
        [stds[0]*stds[1]*corr,           stds[1]**2]] # covariance matrix
table4 = pd.DataFrame(np.random.multivariate_normal(means, covs, 100), columns = ["Mean Anual Temperature", "Total Precipitation"])
display(table4)
```

table4 represents temperature and precipitation data.

It has columns “Mean Annual Temperature” and “Total Precipitation.”

The data is generated using a multivariate normal distribution with specified means, standard deviations, and correlation.

table4:
This table simulates temperature and precipitation data.
It contains two columns:
“Mean Annual Temperature”: Simulated temperature values based on a covariance matrix.
“Total Precipitation”: Simulated precipitation values based on the same covariance matrix.

```python
plt.scatter(table4["Mean Anual Temperature"], table4["Total Precipitation"])
plt.xlabel("Mean Annual Temperature")
plt.ylabel("Total Precipitation")
plt.title("Temperature vs. Precipitation")
plt.show()
```

```python


# Simulate table5
col1 = pd.Series(list(range(1900,2010,10))).repeat(8)
col2 = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]*11
col3 = np.random.uniform(0, 100, 90)
table5 = pd.DataFrame(list(zip(col1, col2, col3)), columns = ["Year", "Landuse", "Cover"])
display(table5)
```

table5 contains historical data for land use cover.

Columns include “Year,” “Landuse,” and “Cover.”

The “Year” column repeats values from 1900 to 2000 in increments of 10.

The “Landuse” column includes the land use categories, and the “Cover” column contains random uniform values between 0 and 100.

table5 (Simulated data with years, land use, and cover):
Bar plot showing the cover values for each land use category:

```python
table5.groupby("Landuse")["Cover"].mean().plot(kind="bar")
plt.xlabel("Land Use Category")
plt.ylabel("Average Cover Value")
plt.title("Average Cover Value by Land Use Category")
plt.show()
```

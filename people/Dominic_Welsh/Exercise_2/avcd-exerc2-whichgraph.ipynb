```python
import pandas as pd
import numpy as np
import random
```

```python
# Simulate var1
# Create empty list to store var1
var1 = []

# Make your random variable not random between uses if you call .seed(24) again
random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)

# Create variable levels, that is a list of landuse types
levels = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]

# Loop is needed because we want 100 random selections from levels
# But there is not a sample method that lets us do that, so using a loop instead
# So the loop will make var1 a list of 100 values, randomly selected from levels
for _ in range(100): # why is a loop needed?
    var1 += random.sample(levels, 1) # var1.append(random.sample(levels, 1)) would also work
```

```python
# Simulate var2
np.random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)

# Generates 100 random values between 0 and 100
var2 = np.random.uniform(0, 100, 100)
```

```python
# Simulate table1
# Creating a Data Frame from var1 (list of landuse types), summing up how many times each value appears
# So it's a table with rows equal to length of var1
table1 = pd.DataFrame(var1).value_counts(sort=True)
# Rename the first column to landuse
table1 = table1.rename_axis("landuse")
# Create a second column named Frequency to more clearly show how many times each value appears
table1 = table1.reset_index(name="Frequency")
display(table1)
```

```python
# Simulate table2
# Create a DataFrame where each landuse is combined with a random number from var2, to give cover area of that land
# It will be a table of 100 rows, each row saying what the landuse type is and the coverage area
table2 = pd.DataFrame(list(zip(var1, var2)), columns = ["landuse", "cover"])
display(table2)

# Note: The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, 
# and then the second item in each passed iterator are paired together, etc. The tuple() function displays a readable version of the result - 
# try running: print(tuple(zip(var1,var2)))
```

```python
# Simulate table3
np.random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
# Create a list of numbers from 1970 to 2021
year = list(range(1970,2021))
# Create a list of random temperature values using normal distribution
# 17 = center of dist, 2 = standard deviation, 51 = number of values (size)
temp = np.random.normal(17,2,51)
# Create a table combining each year with a temp value
# Will be a table of 51 rows having 3 column: #, Year, Temperature
table3 = pd.DataFrame(list(zip(year, temp)), columns = ["Year", "Temperature"])
display(table3)
```

```python
# Simulate table4
# Create an array with the values 16 and 21
xx = np.array([16,21])
# Create an array with the values 300 and 1200
yy = np.array([300, 1200])
# Create a list with the means of two arrays
means = [xx.mean(), yy.mean()]
# Create a list of the standard deviations of the arrays, each divided by 3
stds = [xx.std() / 3, yy.std() / 3]
# Correlation value
corr = -0.7 # correlation
# Covariance matrix, shows the variances in [0,0] and [1, 1]
# and the other 2 numbers show the covariance between xx and yy
covs = [[stds[0]**2          , stds[0]*stds[1]*corr], 
        [stds[0]*stds[1]*corr,           stds[1]**2]] # covariance matrix
# Create a list of 100 random pair of numbers generated from the means and covariance matrix,
# then create a table displaying those pairs as Mean Anual Temperature and Total Precipitation
table4 = pd.DataFrame(np.random.multivariate_normal(means, covs, 100), columns = ["Mean Anual Temperature", "Total Precipitation"])
display(table4)
```

```python
# Simulate table5
# Create a list that has numbers starting at 1900 and going to 2010, jumping by 10 each time.
# Then create a Series, where each value of the list is repeated 8 times
col1 = pd.Series(list(range(1900,2010,10))).repeat(8)
# Create a list that has the following values repeated 11 times
col2 = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]*11
# Create a random uniform array with numbers from 0 to 100, with 90 values in it
col3 = np.random.uniform(0, 100, 90)
# Create a table where the 3 variables (serie, list, array) are zipped together
table5 = pd.DataFrame(list(zip(col1, col2, col3)), columns = ["Year", "Landuse", "Cover"])
display(table5)
```

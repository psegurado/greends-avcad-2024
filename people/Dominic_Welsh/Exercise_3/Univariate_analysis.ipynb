## Descriptive statistics

As an example we will work on a subset of a database that resulted from integrating information from several river fish biomonitoring programmes accross Europe. This subset includes data for some Mediterranean countries. Each case (rows) corresponds to a fish sampling point. Variables (columns) includes coordinates, country and catchment identifiers, local scale environmental variables, climatic variables, human pressures and fish presence/absence data.

When working with a new dataset, one of the most useful things to do is to begin to visualize the data. By using tables, histograms, box plots, and other visual tools, we can get a better idea of what the data may be trying to tell us, and we can gain insights into the data that we may have not discovered otherwise.

We will be going over how to perform some basic visualisations in Python, and, most importantly, we will learn how to begin exploring data from a graphical perspective.

# **Strip Plots**

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(8, 8),
                        gridspec_kw={'wspace': 0.08, 'hspace': 0.2})

sns.stripplot(gn_df, y='temp_ann', ax=axes[0, 0])
sns.stripplot(minho_df, y='temp_ann', ax=axes[0, 1])
sns.stripplot(canta_df, y='temp_ann', ax=axes[1, 0])
sns.stripplot(douro_df, y='temp_ann', ax=axes[1,1])

axes[0, 0].set_title('Galiza-Norte')
axes[0, 1].set_title('Minho')
axes[1, 0].set_title('Cantabrica')
axes[1, 1].set_title('Douro')

plt.show()
```

# **Histograms**

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(8, 8,),
                        gridspec_kw={'wspace': 0.08, 'hspace': 0.2})

sns.histplot(
    gn_df["temp_ann"],
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
    ax=axes[0, 0]
)
sns.histplot(
    minho_df["temp_ann"],
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
    ax=axes[0, 1]
)
sns.histplot(
    canta_df["temp_ann"],
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
    ax=axes[1, 0]
)
sns.histplot(
    douro_df["temp_ann"],
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
    ax=axes[1, 1]
)

axes[0, 0].set_title('Galiza-Norte')
axes[0, 1].set_title('Minho')
axes[1, 0].set_title('Cantabrica')
axes[1, 1].set_title('Douro')

plt.show()
```

# **Boxplots**

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(8, 8,),
                        gridspec_kw={'wspace': 0.08, 'hspace': 0.2})

sns.boxplot(
    x=gn_df["temp_ann"],
    ax=axes[0, 0]
)
sns.boxplot(
    x=minho_df["temp_ann"],
    ax=axes[0, 1]
)
sns.boxplot(
    x=canta_df["temp_ann"],
    ax=axes[1, 0]
)
sns.boxplot(
    x=douro_df["temp_ann"],
    ax=axes[1, 1]
)

axes[0, 0].set_title('Galiza-Norte')
axes[0, 1].set_title('Minho')
axes[1, 0].set_title('Cantabrica')
axes[1, 1].set_title('Douro')

plt.show()
```

### Comments of the plots:


1. The histogram makes it easier to size the temperature concentration values but it can be more complicated to size the extreme values and their interpretation.


2. The boxplot on the other hand shows a general idea of the distribution of the data and clearly shows the presence of outliers in some samples with low temperatures. But it lacks detail in showing this distribution compared to the histogram.


3. The strip plot shows well the distribution of data and outliers, the disadvantage is that it is limited to see important aspects in high density sectors.

```python
# Sample sizes
sample_sizes = [10, 50, 100, 150, 200, 250, 300, 500, 1000]

# Number of random samples for each sample size
num_samples = 1000

# Calculate mean temperature for each sample size and plot in separate windows
for size in sample_sizes:
    temp_means = []
    for _ in range(num_samples):
        # Sample with replacement
        sample = np.random.choice(df2['temp_ann'], size=size, replace=False)
        # Calculate mean temperature of the sample
        temp_means.append(np.mean(sample))
    
    # Plotting
    plt.figure(figsize=(8, 6))
    sns.histplot(temp_means, kde=True, stat='density', alpha=0.6)

    # Plot the true population mean as a vertical line (if known)
    true_population_mean = np.mean(df2['temp_ann'])
    plt.axvline(x=true_population_mean, color='r', linestyle='--', label='Population Mean')

    plt.title(f'Distribution of Sample Means for Sample Size: {size}')
    plt.xlabel('Mean Temperature')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()
```

As an example we will work on a subset of a database that resulted from integrating information from several river fish biomonitoring programmes accross Europe. When working with a new dataset, one of the most useful things to do is to begin to visualize the data. We will be going over how to perform some basic visualisations in Python, and, most importantly, we will learn how to begin exploring data from a graphical perspective.

## Plots secctions

####**Plot Galiza-Norte**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar el dataframe
df_port = df2[df2['Catchment_name'] == 'Galiza-Norte']

# Crear una figura y ejes
fig, axes = plt.subplots(3, 1, figsize=(6, 10))

# Histograma con kde
sns.histplot(
    df_port["temp_ann"],
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
    ax=axes[0]  # Primer eje
).set_title("Histogram of temperature in the catchment of Galiza-Norte", fontsize=12)

# Here the histogram makes it easier to size the temperature concentration values to be in the 14 degree range but it can be
# more complicated to size the extreme values and their interpretation.


# Boxplot
sns.boxplot(
    x=df_port["temp_ann"],
    ax=axes[1]  # Segundo eje
).set_title("Box plot of the annual temperature in the catchment of Galiza-Norte", fontsize=12)

# The boxplot on the other hand shows a general idea of the distribution of the data and clearly shows the presence of outliers in some samples with low temperatures.
# But it lacks detail in showing this distribution compared to the histogram.

# Stripplot
sns.stripplot(
    y=df_port["temp_ann"],
    ax=axes[2]  # Tercer eje
).set_title("Strip plot of annual temperature of the catchment of Galiza-Norte", fontsize=12)

# The strip plot shows well the distribution of data and outliers, the disadvantage is that it is limited to see important aspects in high density sectors.
# (top left sector).

# Ajust the space between plots
plt.tight_layout()

# Show the plots
plt.show()
```

#### Plot samples section

```python
import random
import matplotlib.pyplot as plt
import seaborn as sns

# Bucle para seleccionar valores aleatorios
for i in range(100, 1001, 100):
    # Tomar una muestra aleatoria del DataFrame
    df_random = df.sample(i)

    # Crear una nueva figura y ejes para cada iteraci√≥n
    fig, ax = plt.subplots(figsize=(8, 6))

    # Trazar el histograma con kde
    sns.histplot(
        df_random["temp_ann"],
        kde=True,
        stat="density",
        kde_kws=dict(cut=3),
        alpha=.4,
        edgecolor=(1, 1, 1, 0.4),
        ax=ax
    ).set_title(f"Histogram of temperature for {i} samples", fontsize=12)

    # Mostrar el histograma correspondiente a la muestra actual
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Lista de tama√±os de muestra
sample_sizes = list(range(100, 1001, 100))

# Lista para almacenar las medias de temp_ann de cada muestra
mean_temps = []

# Realizar muestreos aleatorios y calcular la media de temp_ann
for sample_size in sample_sizes:
    # Tomar una muestra aleatoria del DataFrame
    df_random = df.sample(sample_size, replace=True)
    # Calcular la media de temp_ann para la muestra
    sample_mean = df_random['temp_ann'].mean()
    mean_temps.append(sample_mean)

# Visualizaci√≥n
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, mean_temps, marker='o', linestyle='-')
plt.xlabel('Sample size')
plt.ylabel('Mean of the anual temperature')
plt.title('Variation of Anual Temperature mean with sample size')
plt.grid(True)
plt.show()

```

### Extration with variable mean.

```python
import random
import matplotlib.pyplot as plt

# Lista de tama√±os de muestra
sample_sizes = list(range(100, 1001, 100))

# Lista para almacenar las medias de temp_ann de cada muestra
mean_temps = []

# N√∫mero de iteraciones (muestreos aleatorios)
num_iterations = 1000

# Realizar muestreos aleatorios y calcular la media de temp_ann
for sample_size in sample_sizes:
    sample_means = []
    for _ in range(num_iterations):
        # Tomar una muestra aleatoria del DataFrame
        df_random = df.sample(sample_size, replace=True)
        # Calcular la media de temp_ann para la muestra
        sample_mean = df_random['temp_ann'].mean()
        sample_means.append(sample_mean)
    # Calcular la media de las medias de muestra
    mean_temp = sum(sample_means) / len(sample_means)
    mean_temps.append(mean_temp)

# Visualizaci√≥n
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, mean_temps, marker='o', linestyle='-')
plt.xlabel('Sample size')
plt.ylabel('Mean of the anual temperature')
plt.title('Variation of Anual Temperature mean with sample size')
plt.grid(True)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import seaborn as sns

# Lista de tama√±os de muestra
sample_sizes = list(range(100, 1001, 100))

# N√∫mero de iteraciones (muestreos aleatorios)
num_iterations = 1000

# Realizar muestreos aleatorios y calcular la media de temp_ann
for sample_size in sample_sizes:
    sample_means = []
    for _ in range(num_iterations):
        # Tomar una muestra aleatoria del DataFrame
        df_random = df.sample(sample_size, replace=True)
        # Calcular la media de temp_ann para la muestra
        sample_mean = df_random['temp_ann'].mean()
        sample_means.append(sample_mean)

    # Calcular la media de las medias de muestra
    mean_temp = sum(sample_means) / len(sample_means)

    # Crear una nueva figura y ejes para cada muestra
    plt.figure(figsize=(8, 6))

    # Trazar el histograma de las medias de temp_ann para el tama√±o de muestra actual
    sns.histplot(sample_means, kde=True, stat="density", alpha=0.6)
    plt.title(f'Distribution of Temperature Means for Sample Size: {sample_size}')
    plt.xlabel('Mean Temperature')
    plt.ylabel('Density')

    # A√±adir una l√≠nea vertical para la media de las medias de temperatura
    plt.axvline(x=mean_temp, color='r', linestyle='-', label='Mean')

    plt.legend()
    plt.grid(True)
    plt.show()
```

# **Part 4 - Dom**

```python
import numpy as np

# Sample sizes
sample_sizes = [10, 50, 100, 150, 200, 250, 300, 500, 1000]

# Number of random samples for each sample size
num_samples = 1000

# Calculate mean temperature for each sample size and plot in separate windows
for size in sample_sizes:
    temp_means = []
    for _ in range(num_samples):
        # Sample with replacement
        sample = np.random.choice(df2['temp_ann'], size=size, replace=False)
        # Calculate mean temperature of the sample
        temp_means.append(np.mean(sample))

    # Plotting
    plt.figure(figsize=(8, 6))
    sns.histplot(temp_means, kde=True, stat='density', alpha=0.6)

    # Plot the true population mean as a vertical line (if known)
    true_population_mean = np.mean(df2['temp_ann'])
    plt.axvline(x=true_population_mean, color='r', linestyle='--', label='Population Mean')

    plt.title(f'Distribution of Sample Means for Sample Size: {size}')
    plt.xlabel('Mean Temperature')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()
```

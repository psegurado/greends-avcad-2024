```python
import pandas as pd
import numpy as np
import random
```

```python
# Simulate var1
var1 = [] # Crear una lista vacia
random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
levels = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]
for _ in range(100): # why is a loop needed? Para poder iterar eentre los diferentes niveles e ir a√±adiendo uno a uno a lista hasta alcanzar los 100 valores.
    var1 += random.sample(levels, 1) # var1.append(random.sample(levels, 1)) would also work
    
print(var1)
```

```python
# Simulate var2
np.random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice) # Create random numbers with initial seed of 24
var2 = np.random.uniform(0, 100, 100) # Create 100 random numbers with interval of 0-100, with a normal distribution.

print(var2)
```

```python
# Simulate table1
table1 = pd.DataFrame(var1).value_counts(sort=True) # Create a data frame sorted by the # of counts
table1 = table1.rename_axis("landuse") # rename the colunm 
table1 = table1.reset_index(name="Frequency") # rename the colunm 

print(table1)
```

```python
# Simulate table2
table2 = pd.DataFrame(list(zip(var1, var2)), columns = ["landuse", "cover"])

print(table2)
```

```python
# Simulate table3
np.random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
year = list(range(1970,2021)) # Create a list with values in the range of 1970-2021
temp = np.random.normal(17,2,51) #  Create a list with values of 17 = mean, 2 standard deviation and 51 years
table3 = pd.DataFrame(list(zip(year, temp)), columns = ["Year", "Temperature"]) # Create a dataframe

print(table3)
```

```python
# Simulate table4
xx = np.array([16,21]) # Create a array of mean temp
yy = np.array([300, 1200]) # Create a array of total prep
means = [xx.mean(), yy.mean()]  # Simulate/calculate the means values
stds = [xx.std() / 3, yy.std() / 3] # Calculate the standard desviation
corr = -0.7 # correlation # The correlations between temp and total prep will be of -0.7
covs = [[stds[0]**2          , stds[0]*stds[1]*corr], 
        [stds[0]*stds[1]*corr,           stds[1]**2]] # covariance matrix of the co-variance
table4 = pd.DataFrame(np.random.multivariate_normal(means, covs, 100), columns = ["Mean Anual Temperature", "Total Precipitation"])
# Create the data frame with a multivariate normal distribution and create 100 values
print(table4)
```

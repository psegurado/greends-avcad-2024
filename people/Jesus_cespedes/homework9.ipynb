```python
import pandas as pd
import zipfile
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots  
df = pd.read_csv('EFIplus_medit.zip',compression='zip', sep=";")
df = df.dropna() # drops rows when at least one element is a missing value
df.info()
df.head()
```

```python
df = df[(df['Catchment_name']=='Douro') | 
    (df['Catchment_name']=='Tejo')]
```

```python
# Seleccionar las variables de interés
variables = ['Altitude', 'Actual_river_slope', 'Elevation_mean_catch', 'prec_ann_catch', 'temp_ann', 'temp_jan', 'temp_jul']
```

```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Filtrar los datos para seleccionar las cuencas del Duero y el Tajo
df = df[(df['Catchment_name'] == 'Douro') | (df['Catchment_name'] == 'Tejo')]

# Seleccionar las variables de interés
variables = ['Altitude', 'Actual_river_slope', 'Elevation_mean_catch', 'prec_ann_catch', 'temp_ann', 'temp_jan', 'temp_jul']

# Iterar sobre cada variable independiente y ejecutar el dendrograma con el método de enlace "single"
for variable in variables:
    # Calcular el dendrograma con el método de enlace "single"
    dendrogram_plot = dendrogram(linkage(df[variable].values.reshape(-1, 1), method='single'))
    
    # Personalizar y mostrar el dendrograma
    plt.title('Dendrograma con Método de Enlace Single - Variable: ' + variable)
    plt.xlabel('Objetos')
    plt.ylabel('Distancia')
    plt.show()
```

3. Plot a dendrogram that clusters the environmental variables instead (columns) using average linkage - discuss how this would help to select variables for e.g. a regression-based analysis

```python
df = df[['Altitude', 'Actual_river_slope', 'Elevation_mean_catch', 'prec_ann_catch', 'temp_ann', 'temp_jan', 'temp_jul']]

df_t = df.T

df_t.index
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

df = df[['Altitude', 'Actual_river_slope', 'Elevation_mean_catch', 'prec_ann_catch', 'temp_ann', 'temp_jan', 'temp_jul']]

df_t = df.T # El T lo que hace es transponer los datos. 

df_t.index

# Obtener los nombres de las cuencas (sitios) como etiquetas
labels = df_t.index.tolist()

# Tramar el dendrograma
dendrogram_plot = dendrogram(linkage(df_t, method='average'), labels=labels)
plt.title('Dendrograma de Enlace Promedio (Average) - Variables Ambientales')
plt.xlabel('Sitio')
plt.ylabel('Distancia')
plt.xticks(rotation='vertical')
plt.show()
```

The dendrogram is a tool that allows visualizing the hierarchy in the grouping of climatic variables. It is observed that the group of climatic variables, such as temp_jul, temp_ann and temp_jan, are grouped together with the non-climatic variable 'Actual_river_slope', with a low separation distance. On the other hand, topographic variables such as Altitude and average watershed elevation form another group, with a larger distance between them, as seen in the dendrogram. Subsequently, this group is joined to the annual precipitation variable of the watershed.

This representation allows us to identify which variables provide similar information or have very similar behaviors. In the context of a regression, these redundant variables can contribute to multicollinearity.

```python
import pandas as pd
import zipfile
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots
```

```python
df = pd.read_csv('EFIplus_medit (1).zip',compression='zip', sep=";")
print(df)
df.head(10)
df.info()
print(list(df.columns))
```

```python
# clean up the dataset to remove unnecessary columns (eg. REG)
df.drop(df.iloc[:,5:15], axis=1, inplace=True)
print(df)
```

```python
# let's rename some columns so that they make sense
df.rename(columns={'Sum of Run1_number_all':'Total_fish_individuals'}, inplace=True) # inplace="True"
```

```python
# for sake of consistency, let's also make all column labels of type string
df.columns = list(map(str, df.columns))
# Check data types
pd.options.display.max_rows = 154 # maximum number of rows displayed.
df.dtypes
# Number of values per variable
df.count()
```

```python
# Number of missing values (NaN) per variable
df.isnull().sum()
```

```python
df2 = df.dropna(how='all') # drops rows when all elements are missing values
df2.info()
```

```python
df2 = df.dropna(how='all', axis=1) # drops columns when at least one element is a missing value
df2.info()
```

```python
df2 = df.dropna() # drops rows when at least one element is a missing value
df2.info()
```

```python
catchments = ['Minho', 'Douro', 'Galiza-Norte', "Cantabrica"]
df_selected = df[df['Catchment_name'].isin(catchments)]

fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Gráfico 1: Strip plot
sns.stripplot(data=df_selected, x="Catchment_name", y='temp_ann', ax=axs[0])
axs[0].set_title('Strip Plot of Annual Mean Temperature')

# Gráfico 2: Histograma
sns.histplot(data=df_selected, x="temp_ann", y="Catchment_name", ax=axs[1], kde=True)
axs[1].set_title('Histogram of Annual Mean Temperature')

# Gráfico 3: Box plot
sns.boxplot(data=df_selected, x="Catchment_name", y='temp_ann', ax=axs[2])
axs[2].set_title('Box Plot of Annual Mean Temperature')

plt.tight_layout()
plt.show()
```

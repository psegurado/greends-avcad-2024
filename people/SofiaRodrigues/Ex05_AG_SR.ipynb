Standardize , using z score, the "Mean Annual Temperature” (temp_ann), calculate the new mean , SD and 95% confidence interval , and plot the histogram

```python
import numpy as np
import pandas as pd
import scipy.stats as sts
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the dataset
data = pd.read_csv('EFIplus_medit.zip',compression='zip', sep=";")
```

```python
# Assuming Temp_ann is a column in your dataset, replace 'Temp_ann' with the actual column name
temp_ann = data['temp_ann']

# Standardize the Temp_ann variable using z-score3.
temp_ann_zscore = (temp_ann - temp_ann.mean()) / temp_ann.std()

# Calculate the new mean, standard deviation, and 95% confidence interval
new_mean = temp_ann_zscore.mean()
new_std = temp_ann_zscore.std()
conf_interval = sts.norm.interval(0.95, loc=new_mean, scale=new_std)

print("New Mean:", new_mean)
print("New Standard Deviation:", new_std)
print("95% Confidence Interval:", conf_interval)

# Plot the histogram of standardized Temp_ann variable
sns.histplot(temp_ann_zscore, kde=True)
plt.xlabel('Z-score of Mean Annual Temperature')
plt.ylabel('Frequency')
plt.title('Histogram of Standardized Mean Annual Temperature')
plt.show()
```

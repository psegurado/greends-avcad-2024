```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Load the dataset, specifying decimal separator as dot
data = pd.read_csv('EFIplus_medit.zip', compression='zip', sep=";")

# Extract relevant columns
temp_ann_col = 'temp_ann'
brown_trout_col = 'Salmo trutta fario'

if temp_ann_col in data.columns and brown_trout_col in data.columns:
    # Histogram
    plt.figure(figsize=(10, 6))
    plt.hist(data[temp_ann_col], bins=20, color='blue', alpha=0.5)
    plt.title('Histogram of Mean Annual Temperature')
    plt.xlabel('Mean Annual Temperature (¬∞C)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

else:
    print(f"Columns '{temp_ann_col}' or '{brown_trout_col}' not found in the dataset.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro

# Load the dataset, specifying decimal separator as dot
data = pd.read_csv('EFIplus_medit.zip', compression='zip', sep=";")

# Extract the 'Actual_river_slope' column
actual_river_slope = data['Actual_river_slope']

# Visualization: Histogram and Q-Q plot
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(actual_river_slope, kde=True, color='blue')
plt.title('Histogram of Actual River Slope')

plt.subplot(1, 2, 2)
stats.probplot(actual_river_slope, dist="norm", plot=plt)
plt.title('Q-Q Plot of Actual River Slope')

plt.tight_layout()
plt.show()

# Hypothesis Testing: Shapiro-Wilk Test
alpha = 0.05
statistic, p_value = shapiro(actual_river_slope)

print(f"Shapiro-Wilk Test Statistic: {statistic}")
print(f"P-value: {p_value}")

if p_value > alpha:
    print("The Actual River Slope follows a normal distribution (failed to reject H0)")
else:
    print("The Actual River Slope does not follow a normal distribution (reject H0)")
```
